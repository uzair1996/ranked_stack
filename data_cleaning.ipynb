# removing symbol sqrt{2} to √2

```python
import pandas as pd
import re

# Assuming you have already loaded the original dataset as `df`
# Load the dataset if you haven't:
# df = pd.read_csv("path_to_your_file.csv")

# Functions for cleaning text
def remove_html_tags(text):
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text)

def latex_to_symbol(text):
    replacements = {
        r'\sqrt{2}': '√2',
        # You can add more replacements as needed
    }
    for latex, symbol in replacements.items():
        text = text.replace(latex, symbol)
    return text

def clean_text(text):
    text = remove_html_tags(text)
    text = latex_to_symbol(text)
    text = text.replace('$', '')  # remove $ symbols
    return text

# Apply cleaning functions to relevant columns
df['Question'] = df['Question'].apply(clean_text)
df['Answer'] = df['Answer'].apply(clean_text)

# Save the cleaned dataframe to a CSV file
df.to_csv('cleaned_output3.csv', index=False)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt



# Calculate correlation
correlation = df1[['Score', 'ViewCount']].corr()

# Setting a Seaborn theme
sns.set_theme(style="white")

# Defining a color palette
cmap = sns.diverging_palette(220, 20, as_cmap=True)

plt.figure(figsize=(8, 6))

# Drawing the heatmap
ax = sns.heatmap(correlation, annot=True, cmap=cmap, linewidths=.5, linecolor='black', fmt=".2f", annot_kws={'size': 14}, vmin=-1, vmax=1, center=0, square=True, cbar_kws={"shrink": 0.75})

# Set tick label size
ax.tick_params(labelsize=12)

# Title
plt.title('Correlation between Score and ViewCount', fontsize=18)
plt.tight_layout()
plt.show()
```
