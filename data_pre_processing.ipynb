# post history xml to csv

```python
import xml.etree.ElementTree as ET
import csv

# Load the XML data
tree = ET.parse('/Users/uzairpachhapure/csu/Assignments/sem_4/project_final/math.stackexchange.com/PostHistory.xml')
root = tree.getroot()

# Create a CSV file for writing
with open('/Users/uzairpachhapure/csu/Assignments/sem_4/project_final/data_csv/post_history.csv', mode='w', newline='') as csv_file:
    csv_writer = csv.writer(csv_file)

    # Write the header row
    csv_writer.writerow(['Id','PostHistoryTypeId' ,'PostId','RevisionGUID','CreationDate','UserId','Text'])

    # Iterate through XML data and write to CSV
    for row in root.findall('row'):
        id = row.get('Id')
        post_history_type_id = row.get('PostHistoryTypeId')
        post_id = row.get('PostId')
        revision_gui_id = row.get('RevisionGUID')
        creation_date = row.get('CreationDate')
        user_id = row.get('UserId')
        text = row.get('Text')

        csv_writer.writerow([id, post_history_type_id,post_id, revision_gui_id, creation_date,  user_id, text ])

print("Conversion completed. CSV file saved as 'post_history.csv'")
```

```python
import pandas as pd
import time

# Start the timer
start_time = time.time()

# Load the dataset
df = pd.read_csv('/Users/uzairpachhapure/csu/Assignments/sem_4/project_final/data_csv/posts.csv')

# Prepare a list to store our dictionaries before creating DataFrame
data_to_append = []

for index, row in df.iterrows():
    # If the current row's PostTypeId is 1 (indicating a question)
    if row['PostTypeId'] == 1:
        question_body = row['Body']
        question_score = row['Score']
        question_viewcount = row['ViewCount']
        question_creationdate = row['CreationDate']

        # Find the accepted answer using AcceptedAnswerId
        if not pd.isnull(row['AcceptedAnswerId']):
            answer_row = df[df['Id'] == row['AcceptedAnswerId']].iloc[0] if not df[df['Id'] == row['AcceptedAnswerId']].empty else None
            if answer_row is not None:
                answer_body = answer_row['Body']
                answer_score = answer_row['Score']
                # Assumes answer view count is the same as question's, as typically answers do not have separate view counts
                answer_viewcount = question_viewcount
                answer_creationdate = answer_row['CreationDate']
            else:
                answer_body = ''
                answer_score = pd.NA
                answer_viewcount = pd.NA
                answer_creationdate = pd.NA
        else:
            answer_body = ''
            answer_score = pd.NA
            answer_viewcount = pd.NA
            answer_creationdate = pd.NA

        # Append the data as a dictionary to the list
        data_to_append.append({
            'Question': question_body,
            'QuestionScore': question_score,
            'QuestionViewCount': question_viewcount,
            'QuestionCreationDate': question_creationdate,
            'Answer': answer_body,
            'AnswerScore': answer_score,
            'AnswerViewCount': answer_viewcount,
            'AnswerCreationDate': answer_creationdate,
            'OwnerUserId': row['OwnerUserId'],
            'LastEditorUserId': row.get('LastEditorUserId', pd.NA),  # Use .get to avoid KeyError
            'LastEditDate': row.get('LastEditDate', pd.NA),
            'Title': row['Title']
        })

# Create DataFrame from the list of dictionaries
result = pd.DataFrame(data_to_append)

# Save the result to a new CSV file
result.to_csv('output_with_separate_columns.csv', index=False)

# End the timer and print the execution time
end_time = time.time()
print(f"Execution time: {end_time - start_time} seconds")

# Display the first few rows to verify
result.head()
```
